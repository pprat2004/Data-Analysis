```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data_jobs.csv')
```

```python
df = df.dropna(subset = 'salary_year_avg')
df.head(2)
```

```python
salary_median = df.groupby('job_title_short')['salary_year_avg'].median()
salary_median = salary_median.sort_values(ascending = True)
```

```python
plt.barh(salary_median.index,salary_median)
plt.show()
```

```python
salary_mean = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending = True)
salary_mean
```

```python
plt.barh(salary_mean.index,salary_mean)
plt.title('Mean salary by Job Title')
plt.xlabel('Dollars')
plt.show()
```
