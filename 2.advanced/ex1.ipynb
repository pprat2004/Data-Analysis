```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_india = df[df.job_country == 'India']
df_india
```

```python
df_india.head()
```

```python
df_india_pivot = df_india.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_india_pivot
```

```python
df_india_pivot.reset_index(inplace=True)
df_india_pivot['month no'] = pd.to_datetime(df_india_pivot['job_posted_month'],format='%B').dt.month
```

```python
df_india_pivot.sort_values('month no',inplace=True)
df_india_pivot.set_index('job_posted_month',inplace=True)
df_india_pivot.drop(columns='month no',inplace=True)
```

```python
job_titles = ['Data Analyst','Data Engineer','Data Scientist']
df_india_pivot = df_india_pivot[job_titles]
df_india_pivot.plot(kind='line')
plt.ylabel('No of job postings')
plt.xlabel('2023')
plt.title('No of Job listing By Month in India')
```
